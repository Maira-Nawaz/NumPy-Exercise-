Tasks:
Use a for loop to iterate through the list of employees and their work hours.

If an employee works 40 hours or more, they qualify for overtime. Otherwise, they do not.

Add a simple "Overtime" status (Yes/No) for each employee.

Count how many employees qualify for overtime.

Print the final results.

Use a for loop to iterate through the list of employees and their work hours.

```python
employees = ["Alice", "Bob", "Charlie", "David", "Eva"]

work_hours = [35, 45, 50, 20, 40]

for i in employees, work_hours :
  print(i)
```

If an employee works 40 hours or more, they qualify for overtime. Otherwise, they do not.

Add a simple "Overtime" status (Yes/No) for each employee.

```python
overtime_status = []
overtime_count = 0

for i in range(len(employees)):
  if work_hours[i] > 40:
    overtime_status.append("Yes")
    overtime_count += 1
  else:
    overtime_status.append("No")


print("Employee\tWork Hours\tOvertime")
for i in range(len(employees)):
  print(f"{employees[i]}\t{work_hours[i]}\t\t{overtime_status[i]}")

print(f"\nNumber of employees qualified for overtime: {overtime_count}")
```
